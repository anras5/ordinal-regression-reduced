```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate random dataset (10 samples, 2 features)
np.random.seed(42)
data = np.random.randn(10, 2)

# Apply PCA with n_components=2
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(data)

if pca.components_[0, 0] < 0:
    pca.components_[0, :] *= -1
    transformed_data[:, 0] *= -1

if pca.components_[1, 1] < 0:
    pca.components_[1, :] *= -1
    transformed_data[:, 1] *= -1

# Compute the mean of the original data (to position arrows)
mean = np.mean(data, axis=0)
pc1 = pca.components_[0]  # First principal component
pc2 = pca.components_[1]  # Second principal component

# Plot original data
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], color='blue', label="Original Data")
for i, (x, y) in enumerate(data):
    plt.text(x, y, str(i), fontsize=12, ha='right', va='bottom')

# Plot principal component arrows
plt.quiver(mean[0], mean[1], pc1[0], pc1[1], scale=1, color='red', angles='xy', scale_units='xy', width=0.01, label="PC1")
plt.quiver(mean[0], mean[1], pc2[0], pc2[1], scale=1, color='green', angles='xy', scale_units='xy', width=0.01, label="PC2")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Original Data")
plt.legend()

# Plot transformed data
plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], color='red', label="PCA Transformed Data")
for i, (x, y) in enumerate(transformed_data):
    plt.text(x, y, str(i), fontsize=12, ha='right', va='bottom')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Transformed Data")
plt.legend()

plt.tight_layout()
plt.show()
```
